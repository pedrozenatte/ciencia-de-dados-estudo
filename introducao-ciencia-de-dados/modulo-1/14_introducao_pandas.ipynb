# A estrutura de dados básica

A série é uma das principais estruturas de dados do Pandas. Podemos pensar na 'série' como um cruzamento entre lista e dicionário. Os itens são todos armazenados em um pedido e há etiquetas com as quais você pode recuperá-los. 

```python
# Importando a biblioteca Pandas
import pandas as pd
```

```python
# Podemos criar uma série passando uma lista de valores. Quando fazemos isso, o Pandas atribui automaticamente um índice começando com zero e define o nome da série como None.

# Vamos fazer uma lista com três alunos

estudantes = ['Alice,', 'Jack', 'Molly']

# Agora, chamamos a função de série do Pandas e passamos para os alunos.
pd.Series(estudantes)
```

Vemos aqui que o Pandas identifica automaticamente o tipo de dados na série. Os valores são indexados com números inteiros, começando com zero. 

```python
# Veremos que se passarmos uma lista de números inteiros, o Pandas define otipo como int64. Por trás, o Pandas armazena os valores da série em uma matriz digitada usando a biblioteca NumPy, o que oferece uma velocidade significativa ao processar dados em comparação com as listas tradicionais do Python.

numeros = [1, 2, 3]

# Transformando em uma série. 
pd.Series(numeros)
```

```python
# Existem alguns detalhes de digitação para o desempenho que é importante conhecer. O mais importante é como o NumPy e Pandas lidam com os dados perdidos. 

# Em Python, temos o tipo None para indicar falta de dados. Porém, None não um valor numérico nem um valor "vetorizável", na verdade None é um objeto Python especial. 
# Tanto o NumPy quanto Pandas trabalham com arrays vetorizados (usufruem de operações vetorizadas), os quais precisam de um tipo único (dtype) e de operações rápidas em blocos. Dessa forma, nasce um conflito. Como Pandas resolve isso? 
# Por baixo, o Pandas faz uma conversão de tipo para nós. Se criarmos uma lista de str e tivermos um elemento faltante, um tipo None, o Pandas analisa isso como uma coluna de determinado tipo, mas com dados faltantes, e, nesse caso, ele troca o tipo None para --> NaN (marcados padrão de dado ausente no Pandas) para não quebrar operações vetorizadas. 

# Vemos criar uma nova lista de estudantes, mas vamos definí-la como None.
estudantes2 = ['Alice', 'Jack', None]
pd.Series(estudantes2)
```

```python
# Vamos criar uma lista de inteiros e colocamos o tipo None, o Pandas automaticamente a converterá em um valor especial de ponto flutuante designado como NaN.
numeros2 = [1, 2, None]
pd.Series(numeros2)
```

Perceba que mesmo tendo inteiros, o Pandas generalizou para Float e utilizou o NaN. Por que não deixar isso como um número inteiro? <br>
R: Justamente porque o Pandas considera o NaN, nesse caso, como um tipo float, e assim é utilizado o tipo mais geral, e dessa forma, converteu todos os valores para floats.

##### OBS: Se tivermos uma coluna com apenas Inteiros (int) e ao transformar em série temos o tipo float, MUITO PROVAVELMENTE estará com dados faltantes. 

```python
# É importante enfatizar que None e NaN podem ser utilizado da mesma forma para indicar dados ausentes. CONTUDO, PORÉM, TODAVIA, ENTRETANTO, por baixo dos panos, eles NÃO são representados pelo Pandas da mesma forma. 

# NaN não é equivalente a None. 
import numpy as np      # Biblioteca que permite criarmos o tipo NaN
np.nan == None
```

```python
# Também não podemos fazer o teste de qualidade do NaN sozinho
np.nan == np.nan
```

```python
# Para verificar o tipo, temos que usar funções especiais, como utilizamos a isinstance para verificar os próprios tipos do Python. 

np.isnan(np.nan)
```

Lembre-se que quando vemos um NaN, o seu significado é semelhante a None, mas NaN é um valor numérico e é tratado de forma diferente por motivos de EFICIÊNCIA. 

OBS: NaN é um valor especial que o NumPy criou.
