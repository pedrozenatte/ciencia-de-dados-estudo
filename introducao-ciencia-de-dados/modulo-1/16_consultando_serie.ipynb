# Consultando uma série

É importante pensar em paralelização ao se envolver na programação de ciência de dados

```python
import pandas as pd
```

```python
# Uma série Pandas pode ser consultada pela posição do índice ou pelo rótulo do índice. Se não for fornecido um índice para a série durante a consulta, a posição e o rótulo são os mesmos valores.
# Para consultar por localização numérica, começando em 0, use o atributo iloc. Para consultar pelo rótulo do índice, pode utilizar o atributo loc.

# Vamos utilizar alunos matriculados em aulas provenientes de um dicionário. 
aulas_estudantes = {'Alice': 'Física', 
                    'Jack': 'Química',
                    'Molly': 'Inglês',
                    'Sam': 'História'}

s = pd.Series(aulas_estudantes)
s
```

```python
# Para esta série, se quisermos ver a quarta entrada, podemos usar o atributo iloc com o parâmetro 3.
s.iloc[3]
```

```python
# Se quisermos ver qual classe Molly tem, usaríamos o atributo loc com um parâmetro de molly.
s.loc['Molly']
```

Mas qual a diferença entre:
- s.loc['Molly']
- s.Molly

A diferença é como o Pandas está acessando o dado. A forma s.loc é a forma mais segura, funciona sempre, com qualquer nome, mesmo se tiver espaço, número, etc. <br>
Agora quando temos um s.atributo, o Pandas permite isso apenas quando o nome do índice é um identificador válido de Python, isto é, não pode ter espaço, não pode começar com número, e não pode ter nome de métodos, como mean. 

```python
# Perceba que não podemos acessar um valor pelo método de indexação, como s[0], apenas se o rótulo for 0

estudantes = ['Alice,', 'Jack', 'Molly']
serie = pd.Series(estudantes)
serie[0]
```

```python
# O que acontece se seu índice for uma lista de números inteiros? 
# O Pandas não pode determinar automaticamente se você pretende consultar por posição de índice ou rótulo de índice, então é necessário ter cuidado ao usar, ou seja, a opção mais segura é ser bem explícito (utilizando iloc e loc). 

codigo_classe = {99: 'Fisica', 100: 'Química', 101: 'Inglês', 102: 'História'}
s = pd.Series(codigo_classe)
print(s[99])
print(s.loc[99])
print(s.iloc[0])
```
