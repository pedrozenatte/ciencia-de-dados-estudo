# Merging de DataFrames

Em análise de dados com pandas, é comum precisarmos combinar múltiplos DataFrames. Esse processo é conhecido como merging (junção) e está diretamente relacionado aos conceitos de joins em Sistemas de Base de Dados (Banco de Dados Relacional). <br>
Em teoria relacional, cada DataFrame pode ser entendido como uma relação (tabela), e cada linha como uma tupla. A operação de junção ocorre quando combinamos duas tabelas com base em um atributo comum (chave).

Para entender os diferentes tipos de junção, podemos utilizar a analogia com Teoria dos Conjuntos e Diagramas de Venn.

![Venn Diagram](img/merging1.png)



#### Diagrama de Venn e Populações

Considere dois conjuntos:

- Conjunto A: Estudantes
- Conjunto B: Funcionários (Staff)

Existe uma interseção entre esses conjuntos, isto é, indivíduos que são simultaneamente estudantes e funcionários.

No contexto de Base de Dados, isso equivale a duas tabelas distintas que possuem uma coluna em comum (por exemplo, Name ou ID).

`1 - Full Outer Join (União)`

Em Base de Dados, o Full Outer Join retorna todos os registros de ambas as tabelas, independentemente de haver correspondência. <br>
Na Teoria dos Conjuntos, isso corresponde à união.

Interpretação: <br>
Seleciona todos os estudantes e todos os funcionários, incluindo os que não aparecem na outra tabela. <br>
Quando não houver correspondência, os valores da outra tabela aparecem como NULL (ou NaN no pandas).

**CÓDIGO: pd.merge(df_students, df_staff, ***how="outer"***, on="Name")**

`2 - Inner Join (Interseção)`

Em Base de Dados, o Inner Join retorna apenas os registros que possuem correspondência em ambas as tabelas. <br>
Na Teoria dos Conjuntos, isso corresponde à interseção.

Interpretação: <br>
Seleciona apenas os indivíduos que aparecem nas duas tabelas.

**CÓDIGO: pd.merge(df_students, df_staff, ***how="inner"***, on="Name")**

`3 - Left Join`

O Left Join mantém todos os registros da tabela da esquerda e adiciona informações da tabela da direita apenas quando houver correspondência.

**CÓDIGO: pd.merge(df_students, df_staff, ***how="left"***, on="Name")**

`4 - Right Join`

O Right Join é o inverso do Left Join: mantém todos os registros da tabela da direita.

**CÓDIGO: pd.merge(df_students, df_staff, ***how="right"***, on="Name")**

#### Resumo

| Tipo de Join | Teoria dos Conjuntos        | Base de Dados   | pandas        |
| ------------ | --------------------------- | --------------- | ------------- |
| Full Outer   | União                       | FULL OUTER JOIN | `how="outer"` |
| Inner Join   | Interseção                  | INNER JOIN      | `how="inner"` |
| Left Join    | Parte esquerda + interseção | LEFT JOIN       | `how="left"`  |
| Right Join   | Parte direita + interseção  | RIGHT JOIN      | `how="right"` |


```python
# Importando a biblioteca
import pandas as pd

# Vamos criar dois dataframes, um de funcionários e um de estudantes
funcionarios_df = pd.DataFrame([{'Nome': 'Kelly', 'Funcao': 'Diretor de RH',},
                                {'Nome': 'Sally', 'Funcao': 'Ligação do Curso'},
                                {'Nome': 'James', 'Funcao': 'Avaliador'}])

# Vamos indexar esses funcionarios pelo nome:
funcionarios_df.set_index('Nome', inplace = True)

# Dataframe de alunos
estudantes_df = pd.DataFrame([{'Nome': 'James', 'Curso': 'Negócios'},
                           {'Nome': 'Mike', 'Curso': 'Advocacia'},
                           {'Nome': 'Sally', 'Curso': 'Engenharia'}])

# Vamos indexar esses estudantes pelo nome:
estudantes_df = estudantes_df.set_index('Nome')

# Printando
display(estudantes_df.head())
display(funcionarios_df.head())
```

Perceba que há uma sobreposição nesses DataFrames, pois James e Sally são tanto estudantes quanto funcionários, porém Mike e Kelly não.

É importante ressaltar que os dois dataframes são indexados de acordo com o valor em que queremos mesclar, que, nesse caso, é o nome. 

**OBS:** Não é obrigatório que ambos estejam indexados por algo em comum, mas é obrigatório que exista alguma coluna ou índice em comum que sirva como chave de junção.


**OBS sobre o print():** No Jupyter Notebook, quando você escreve apenas df.head() como última linha da célula, o ambiente renderiza automaticamente o DataFrame em formato HTML, exibindo uma tabela organizada e visualmente estruturada. Isso acontece porque o Jupyter utiliza o método interno de representação gráfica do objeto (_repr_html_()). <br>
Por outro lado, quando usamos print(df.head()), o Python força a exibição da representação textual simples do objeto, mostrando a tabela em formato de texto puro no console. <br>
Portanto, para visualizar a tabela formatada, basta chamar o DataFrame diretamente, sem utilizar print(), ou, se você precisa usar dentro de uma função ou no meio da célula, utilizar display(), pois isso força a renderização HTML bonita.

#### **1 - União**

```python
# Se queremos a união deles, chamaremos a função merge(), passando o quadro de dados à esquerda e o quadro de dados à direita e dizendo ao merge que queremos que ele use uma junção externa.
# Queremos usar os índices esquerdo e direito como colunas de união

pd.merge(funcionarios_df, estudantes_df, how = 'outer', left_index = True, right_index = True)
# OBS: O funcionarios_df é a tabela a esquerda, pois está colocado a esquerda. Já o estudantes_df é a tabela a direita, pois foi colocado a direita de funcionarios_df, o que significa que left_index está associado ao funcionarios_df, por exemplo. 
```

Quando fazemos left_index = True, e right_index = True, estamos falando o seguinte para a função:

- Use o índice do DataFrame da esquerda como chave de junção
- e use o índice do DataFrame da direita como chave de junção.

Perceba no resultado, que todos estão listados, e como a Kelly não tem um curso e o Mike não tem uma função, o valor é NaN, representando dados ausentes. 

#### 2 - **Interseção**

```python
# Se quisermos obter a interseção entre apenas as pessoas que são estudantes e funcionários, podemos definir o atributo how como 'inner'.
pd.merge(funcionarios_df, estudantes_df, how = 'inner', left_index = True, right_index = True)
```

Percebe-se que o resultado possui apenas Sally e James, uma vez que apenas eles são tanto funcionários quanto estudantes

#### 3 - **Left Join**

```python
# Aqui temos a adição de conjuntos. 
# Imagine que queremos obter a lista de todos os funcionários, independente de serem estudantes ou não, mas se eles são estudantes, também queremos obter os detalhes dos alunos. Para isso, vamos usar o 'left'. 
# LEMBRANDO QUE o primeiro DataFrame e o df que representa a esquerda, e o segundo é o que representa a direita.

pd.merge(funcionarios_df, estudantes_df, how = 'left', left_index = True, right_index = True)
```

Perceba que o DataFrame resultante possui todos os funcionários, e aqueles que são estudantes também tem os dados de estudantes sendo apresentados.

#### 4 - **Right Join**

```python
# Agora, queremos uma lista de todos os alunos e, se eles também forem funcionários, queremos os dados de funcionários. Para isso, vamos usar o 'righ'.

pd.merge(funcionarios_df, estudantes_df, how = 'right', left_index = True, right_index = True)
```

Perceba que o DataFrame resultante possui todos os estudantes, e aqueles que são funcionários também tem seus dados de funcionários sendo apresentados.

##### OUTRAS FORMAS DE MESCLAR

Se você quiser mesclar usando uma coluna específica como chave, em vez do índice, você usa o parâmetro:

- on = 'nome_da_coluna'

#### Caso 1 — A coluna tem o mesmo nome nos dois DataFrames

```python
# E quisermos colocar a chave de junção com colunas diferentes? Lembrando que não é obrigatório ser a coluna de índice do DataFrame, basta ter uma coluna que ligue a outra (como se fosse uma chave estrangeira)

pd.merge(funcionarios_df, estudantes_df, 
         how='outer', 
         on='Name')
```

#### Caso 2 — A coluna tem nomes diferentes

Exemplo: 
- Em funcionarios_df → 'Nome_funcionario'
- Em estudantes_df → 'Nome_estudante'

```python
pd.merge(funcionarios_df, estudantes_df,
         how = 'outer',
         left_on = 'Nome_funcionario',
         right_on = 'Nome_estudante')
```

#### Caso 3 — Usar índice de um lado e coluna do outro

```python
pd.merge(funcionarios_df, estudantes_df,
         left_index = True,
         right_on = 'Nome')
```

#### Resumo

| Situação              | Parâmetro                              |
| --------------------- | -------------------------------------- |
| Mesma coluna nos dois | `on = 'coluna'`                          |
| Colunas diferentes    | `left_on` e `right_on`                 |
| Usar índice           | `left_index = True` / `right_index = True` |


##### ATENÇÃO: 

Não é possível utilizar o merge com mais de dois DataFrames, então se quiser mesclar vários DataFrames, e necessários ir mesclando o resultado da mescla anterior, ou utilizar a função reduce.
