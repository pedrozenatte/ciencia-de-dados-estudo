# Indexação de DataFrames

Como vimos, tanto Séries como DataFrames podem ter índices aplicados. O índice é basicamente um rótulo de cada linha, e sabemos que linhas correspodem ao eixo (axis) zero. <br>
Os índices podem ser gerados automaticamente quando criamos uma nova série sem um índice e, nesse caso, obtemos um valor numérico, ou eles podem ser definitos explicitamente quando usamos o objeto dicionário e criamos uma nova Série ou quando carregamos arquivo CSV e definimos os parâmetros apropriados. <br>
Outra opção para configuração de um índice é usar a função set_index(). Essa função usa uma lista de colunas como parâmetro e promove tais colunas como índice. 

```python
# Importanto a biblioteca
import pandas as pd
```

```python
# A função set_index() é um processo destrutivo, e não mantém o índice atual. Se você quiser manter o índice atual, precisa criar manualmente uma nova coluna e copiar nela os valores do atributo índice.

df = pd.read_csv('datasets/Admission_Predict.csv', index_col = 0)
df.head()
```

```python
# Vamos supor que não queremos mais como índice do DataFrame o número de serial, mas, em vez disso, a coluna "Chance of Admit", porém vamos manter a coluna "Serial No." para usar depois. Logo, iremos preservar a coluna "Serial No." em uma nova coluna.
# Podemos fazer isso usando o operador de indexação com a str que é o nome da coluna. Usaremos a função set_index para atribuir como índice a coluna "Chance of Admit"

# Copiando os dados da coluna índice para a nova coluna
df['Numero de Série'] = df.index

# Agora podemos definir como o índice uma outra coluna, sem que perdessemos o índice inicial
df.set_index('Chance of Admit ', inplace = True)
df.head()
```

Lembrando que o inplace é para a mudança ocorrer no DataFrame original, uma vez que a função set_index retorna uma cópia, sem alteração do original.

Vemos que criamos um novo índice de uma coluna existente, e o índice tem o mesmo nome da coluna.

```python
# Podemos resetar o índice usando a função reset_index(). Ela promove o índice, tornando-o uma coluna e cria uma coluna índice padrão numerada.
df.reset_index(inplace = True)
df.head()
```

Perceba que "Chance of Admit" é promovida de volta como uma coluna e temos um índice numérico

```python
# Um bom recurso do Pandas é a indexação em multi-nível. Isso é semelhante ao uso de chaves compostas em sistemas de bancos de dados relacionais. 
# Para cria um índice multi-nível nós usamos set_index() e informamos a lista de colunas que estamos interessados

df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
# Por exemplo, nesse conjunto de dados, há dois níveis de sumarização, um que contém dados sumarizados (informações brutas condensadas) de todo o país, e outro que contém dados sumarizados de cada estado.
# Queremos ver a lista de todos os valores únicos em uma determinada coluna. 
# Nesse DataFrame, vemos que os possíveis valores para a coluna "SUMLEV" usando a função unique()

df["SUMLEV"].unique()
```

Perceba que o resultado nos mostra dois valores apenas, ou seja, a coluna "SUMLEV" só tem dois valores possíveis, ou 40 (Estado) ou 50 (Condado).

```python
# Vamos excluir todas as linhas que são resumos no nível estadual e mantenhamos dados de condados. 
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
# Além disso, enquanto esse conjunto de dados for interessante por um número de diferentes razões, vamos reduzir nosso conjunto de dados para ver apenas o total de estimativas populacionais e o número total de nascimentos.

# Podíamos excluir todas as colunas que não queremos, porém, neste caso, como existem muitas, é mais fácil escrever as que queremos manter.
# Podemos começar criando uma lista de nomes das colunas que queremos manter
colunas_matidas = ['STNAME','CTYNAME','BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013',
                   'BIRTHS2014','BIRTHS2015','POPESTIMATE2010','POPESTIMATE2011',
                   'POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']

df = df[colunas_matidas]
df.head()
```

```python
# Os dados do Censo dos EUA possibilitam separar as estimativas populacionais por estado e condado
# Podemos carregar os dados e usando set_index() para combinar valores de estados e de condados.

df.set_index(['STNAME', 'CTYNAME'], inplace = True)
df.head(10)
```

```python
# Como podemos consultar esse DataFrame?
# Por exemplo, vimos que o atributo "loc" do DataFrame pode ter vários argumentos, e ele poderia consultar ambas linhas e colunas.
# Quando usamos multi-index precisamos fornecer os argumentos na ordem do nível que desejamos consultar. Dentro do índice, cada coluna é chamada de 'nível', de modo que a coluna mais externa é o nível zero. 

# Então, se quisermos ver os resultados da população do condado de Washtenaw, no estado de Michigan, o primeiro argumento seria "Michigan" e o segundo "Washtenaw County"
df.loc['Michigan', 'Washtenaw County']
```

```python
# E se quisermos filtrar apenas duas colunas? 
# Nesse caso, precisamos NECESSARIAMENTE passar os níveis entre TUPLAS, para conseguirmos separar as colunas
df.loc[('Michigan', 'Washtenaw County'), ['BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012']]
```

```python
# E se quisermos filtrar mais níveis?
# Então devemos passar uma lista de tuplas
df.loc[[('Michigan', 'Washtenaw County'), ('Alabama', 'Autauga County')], ['BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012']]
```

```python
# Se for o mesmo nível 0 mas diferentes níveis 1, podemos utilizar a tupla, com uma lista dentro da tupla
df.loc[('Alabama', ['Autauga County', 'Baldwin County']), ['BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012']]


# Isso evita ter que ficar repetindo Alabama, por exemplo, df.loc[[('Alabama', 'Autauga County'), ('Alabama', 'Baldwin County')], ['BIRTHS2010', 'BIRTHS2011']]
```
