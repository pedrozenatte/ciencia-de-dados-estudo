# Criando a série Pandas

Como já dito, a Série é uma estrutura de dados do Pandas. Embora a lista possa ser uma forma comum de criar alguns dados, geralmente temos dados com rótulos que desejamos manipular. <br>
Uma série pode ser criada diretamente dos dados do dicionário e, se fizermos isso, o índice será automaticamente atribuído às chaves do dicionário. 

```python
# Importando a bilbioteca
import pandas as pd
```

```python
# Vamos criar um dicionário chamado notas dos estudantes. 
aulas_estudantes = {'Alice': 'Física', 
                    'Jack': 'Química',
                    'Molly': 'Inglês'}

s = pd.Series(aulas_estudantes)
s
```

```python
# A partir da Série ser criada, podemos obter o objeto de índice usando o atributo index. 
s.index
```

```python
# À medida que utilizamos mais o Pandas, notaremos que muitas coisas são implementadas como matrizes numpy e que elas têm o valor dtype definido. Isso vale para os índices, e aqui o Pandas inferiu que estamos usando strings para o índice. 

# Isso é interessante porque o tipo do objeto não é apenas para strings ou floats, é para objetos arbitrários (mais complexos).

# Vamos criar um tipo de dados mais complexo --> uma lista de tuplas.
estudantes = [('Alice', 'Brown'), ('Jack', 'White'), ('Molly', 'Green')]
pd.Series(estudantes)
```

Vemos que cada uma das tuplas é armazenada no objeto da série e o tipo é do tipo object. 

```python
# Também podemos separar a criação do índice dos dados passando um índice como uma lista explicitamente para a série.

s = pd.Series(['Física', 'Química', 'Inglês'], index = ['Alice', 'Jack', 'Molly'])
s
```

```python
# O que acontece se a lista de valores no objeto de índice não estiver alinhada com as chaves em seu dicionário para criar a série?? 
# O Pandas substitui a criação automática para favorecer somente os valores de índice que foi fornecido. Portanto, será ignorado do dicionário todas as chaves que não estão no seu índice e o Pandas adicionará NaN.

# Vamos passar um dicionário de três itens, com alunos e seus cursos
curso_estudante = {'Alice': 'Física', 
                    'Jack': 'Química',
                    'Molly': 'Inglês'}

# Quando criar a serie, vamos pedir apenas um índice com três alunos e excluirei Jack. 
s = pd.Series(curso_estudante, index = ['Alice', 'Molly', 'Sam'])
s
```
