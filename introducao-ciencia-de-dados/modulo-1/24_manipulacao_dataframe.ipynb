# Manipulação de DataFrame

```python
# Importando a biblioteca
import pandas as pd

# Lendo o arquivo
df = pd.read_csv('datasets/presidents.csv')

df.head()
```

```python
# Vamos colocar o index para a coluna '#'
df.set_index('#', inplace = True)
df.head()
```

Temos agumas notas de rodapé na coluna 'born', então vamos tirar essas notas, pois elas podem causar problemas. Além disso, também vamos ajustar os presidentes em primeiro nome e sobrenome.

```python
# 1° Forma: 
# Vamos copiar a coluna de president
df['Primeiro nome'] = df['President']

# Agoram podemos modificar a nova coluna com replace() 
df['Primeiro nome'] = df['Primeiro nome'].replace('\s[\w]*', '', regex = True)
df.head()

```

Isso que fizemos funciona, porém é meio inadequado e é lento, pois tivemos que fazer uma cópia completa da coluna antes e depois atualizar as strings. 

```python
# 2° Forma:
# Forma mais genérica, que é a função apply()

# Removendo a coluna que criamos
del df['Primeiro nome']

# A função apply() em um DataFrame pega alguma função arbitrária que você codificou e a aplica a uma série, a uma única coluna ou ao DataFrame total, de forma iterativa
def separa_nome(linha: pd.Series):
    # Cada chamada dessa função recebe uma linha do DataFrame, a qual é passada como uma Série
    # Vamos extrair o primeiro nome e criar uma nova entrada para a série
    linha['Primeiro nome'] = str(linha['President']).split()[0]
    linha['Sobrenome'] = str(linha['President']).split()[-1]

    return linha

# Agora aplicamos isso ao DataFrame indicando que queremos aplicar nas colunas
df = df.apply(separa_nome, axis = 1)
df.head(10)
```

```python
# 3° Forma: 

# Removendo as colunas que criamos
df.drop(['Primeiro nome', 'Sobrenome'], axis = 1, inplace = True)

# Vamos usar a função extract(), pois ela pega expressões regulares (regex) como entrada, atribuindo o valor desejado de grupos de captura que correspondem às colunas de saída que estamos interessados
nomes = df['President'].str.extract('(?P<Primeiro_nome>^[\w]*)(?:.*\s)(?P<Sobrenome>[\w]*$)')
nomes
```

OBS: Em regex, o nome do grupo não pode ter espaços

```python
# Agora, basta passar isso para o DataFrame
df['Primeiro nome'] = nomes['Primeiro_nome']
df['Sobrenome'] = nomes['Sobrenome']
df.head()
```

Agora vamos olhar para a parte do 'Born'

```python
# Primeiro, vamos nos livrar de tudo que não está no padrão dia, mês, ano
df['Born'] = df['Born'].str.extract('(?P<Data>[\w]{3}\s[\d]{1,2}\,\s[\d]{4})(?:\[.*\])?')
df['Born']
```

OBS: As funções '.str.' são vetorizadas

```python
# Perceba que o tipo dessa coluna é "str", mas o Pandas tem funcionalidades interessantes para lidar com data/hora

# Vamos transformar a str em data/hora 
df['Born'] = pd.to_datetime(df['Born'])
df['Born'].head()
```

Isso faria com que o processamento subsequente nos dados relacionados a datas, como conseguir todos os presidentes nascidos em determinado período, se torne muito mais fácil
