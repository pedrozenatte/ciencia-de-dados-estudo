# Padrões de classes e caracteres (conjunto)

```python
import re
```

```python
# Vamos criar uma série de notas de alunos individuas ao longo de um semestre em um curso em todas as suas tarefas individuais. 

notas = "ACAAAABCBCBAA"

# Se quisermos responder à pergunta, quantos Bs estavam na lista de notas, poderíamos simplesmente usar B como o nosso padrão

re.findall('B', notas)
```

```python
# Se quisermos contar o número de 'A's ou 'B's na lista, não podemos usar A e B, pois isso é usado para corresponder a todos os 'A's seguidos imediatamente por 'B' (será considerado a str "AB"). 

# Para fazer do jeito que queremos, passaremos a string dentro de []
re.findall('[AB]', notas)
```

```python
# Você também pode incluir uma variedade de caracteres ordenados alfanumericamente. Por exemplo, se quisermos nos referir a todos os caracteres minúsculos, podemos usar [a-z] nos colchetes. 

# Vamos criar um regex simples para analisar todas as intâncias em que o aluno recebe um A seguido porum Bou um C. Então: 

re.findall('[A][B-C]', notas) 

# re.findall('A[B-C]', notas) também funciona. 
```

Perceba como o padrão A, B descreve um conjunto de caracteres possíveis que podem ser A ou B, enquanto o padrão A e seguido pelo padrão B-C denotam dois conjuntos de caracteres que devem ter sido combinados consecutivamente.

```python
# Você pode escrever esse padrão usando o operador pipe "|", que significa OU. 

re.findall('AB|AC', notas)
```

```python
# Podemos usar o cursor com o operador set (o operador set no contexto de regex é "[]"") para negar nossos resultados "^". Por exemplo, se quiséssemos analisar apenas as notas que não eram 'A's, faríamos: 

re.findall('[^A]', notas)
```

```python
# Perceba que o símbolo ^ normalmente indica o início da string. Porém, quando ele aparece dentro de um conjunto de caracteres ([]), ele deixa de ser uma âncora e passa a ter outro significado (negação), assim como outros caracteres especiais também perdem seus significados originais.

re.findall('^[^A]', notas)

# Regex é contextual, isto é, o mesmo símbolo pode ter significados diferentes dependendo de onde aparece. Nesse caso, ele deixa de ser âncora.
```

É uma lista vazia porque o regex está dizendo que queremos corresponder a qualquer valor no início da string, que não seja um A, e nossa string começa com um A.
