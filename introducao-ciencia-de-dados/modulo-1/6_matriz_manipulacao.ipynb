# Matriz de manipulação

Podemos fazer muitas coisas em matrizes, como manipulação matemática, adição, subtração, expoentes quadrados e matrizes booleanas, que são valores binários. Também podemos fazer manipulação de matrizes, como transposição de produto, inversa e assim por diante.

```python
import numpy as np
```

```python
# Operadores aritméticos em vetores se aplicam a elementos. 

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a - b
print(c)

c = a * b
print(c)

# Com a manipulação aritimética, podemos converter os dados atuais do jeito que queremos que sejam. 
# Os operadores aritméticos em vetores do NumPy permitem realizar operações element-wise (ou seja, operação realizada individualmente em cada elemento do vetor).
# ATENÇÃO: As dimensoes dos vetores/matrizes precisam ser iguais para funcionar
```

```python
# Precisão de duas casas decimais
%precision 2

# Vamos transformar farenheit em celcius
farenheit = np.random.randint(-50, 100, 15)
farenheit.sort()

celsius = (farenheit - 31) * (5/9)
print(celsius)
```
