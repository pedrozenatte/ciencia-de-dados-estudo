# Valores ausentes

Vimos uma prévia de como o Pandas lida com valores ausentes usando o tipo NaN. <br>
Valores ausentes são muito comuns em atividades de limpeza de dados, e valores ausentes podem existir por vários motivos:

- Pode haver, dentro de uma pesquisa, respostas que o respondente não forneceu, ou seja, uma OMISSÃO. Esse dado ausente, chamamos de **Missing at Random** (ausente aleatoriamente) se houvver outras variáveis que possam ser usadas para prever a variável que está ausente. Se os dados faltantes não possuem relação com nenhuma outra variável chamamos esses dados de **Missing Completely at Random** (desaparecidos completamente de forma aleatória)
- Dados ausentes por não terem sido coletados

```python
# Importando a biblioteca 
import pandas as pd
```

```python
# Pandas é muito bom em detectar valores ausentes diretaemnte de arquivos como CSV.
# Embora a maioria dos valores ausentes geralmente seja formatada como NaN, null, None, NA, às vezes os valores ausentes não são rotulados de forma tão clara, por exemplo, utilizar 99 em categorias binárias para indicar que é um valor ausente.

# A função pandas read_csv() tem um parâmetro chamado "na_values" que nos permite especificar o formato dos valores ausentes. Ele permite que sequências escalares, listas ou dicionários sejam usados.
df = pd.read_csv('datasets/class_grades.csv')
df.head(10)
```

Perceba que existem vários valores ausentes apenas nessas 10 linhas.

```python
# Podemos usar a função .isnull() para criar uma máscara booleana de todo o quadro de dados. Isso vai funcionar por broadcast
mascara = df.isnull()
mascara
```

```python
# Isso pode ser útil para processar linhas com base em determinadas colunas de dados. Outra operação útil é poder eliminar todas as linhas que têm dados ausentes, o que pode ser feito com a função dropna()
df.dropna().head(10)
```

Perceba que algumas linhas foram removidas

```python
# Uma das funções úteis que o Pandas tem para trabalhar com valores ausentes é a função de preenchimento chamada fillna(). Essa função usa vários parâmetros, de modo que podemos passar um único valor, chamado de valor escalar, para transformar todos os dados ausentes em um único valor

# Se quiséssemos preencher todos os valores ausentes com zero, podemos usar fillna()
df.fillna(0, inplace = True)
```

```python
# Também podemos usar o parâmetro na_filter para desativar a filtragem de espaço em branco se o espaço em branco for um valor real de interesse. 
# OBS: Na prática isso é muito raro, pois em dados sem nenhum NAs, passar na_filter = False pode melhorar o desempenho da leitura de um arquivo grande 
# pd.read_csv("arquivo.csv", na_filter = False) 
"""
Quando pedimos isso, estamos falando: Não procure valores ausentes. Leia tudo exatamente como está.”

Ou seja:
Ele não converte strings vazias em NaN
Ele não converte "NA" em NaN
Ele simplesmente carrega os dados como texto bruto
"""

df = pd.read_csv('datasets/log.csv')
df.head(20)
```

```python
# Vamos setar o index com base na coluna 'time'
df = df.set_index('time')

# Vamos ordenar com base no index, pois ao mudar o index, vai ficar desordenado
df = df.sort_index()
df.head(20)
```

```python
# Se observarmos atentamente a saída, perceberemos que o índice não é realmente único. Parece que dois usuários conseguem usar o sistema ao mesmo tempo. 
# Nesse caso, vamos redefinir o índice e usar uma indexação multi-nível sobre pontualidade e o usuário

df.reset_index(inplace = True)
df.set_index(['time', 'user'], inplace = True)
df.sort_index(inplace = True)
df
```

```python
# Agora que temos os dados indexados e classificados adequadamente, podemos preencher os dados ausentes usando ffill. 
# É bom lembrar que, ao lidar com valores ausentes, você pode lidar com colunas indivudais ou conjuntos de colunass projetando-os para que não precise corrigir todos os valores ausentes em um comando. 

df = df.ffill()
df.head()
```

##### Essa função ffill() faz:
- Ele preenche valores ausentes com o último valor válido acima.

Já a fillna() preenche com um valor especificado. 

```python
# Também podemos fazer um preenchimento personalizado para substituir os valores com a função replace(). Ela permite a substituição de várias abordagens, valor para valor, lista, dicionário, regex, então vamos gerar um exemplo simples.
df = pd.DataFrame({'A': [1, 1, 2, 3, 4], 
                   'B': [3, 6, 3, 8, 9],
                   'C': ['a', 'b', 'c', 'd', 'e']})
df
```

```python
# Poderíamos substituir aquele por 100. Então vamos tentar a abordagem de valor para valor 
# Vamos passar na função replace() primeiro o valor que vamos substituir e depois o valor que entrará no lugar.
df.replace(1, 100, inplace = True)
```

```python
# Agora vamos mudar tanto o 100 quanto o 3 para 103 e 300
df.replace([100, 3], [103, 300])
```

```python
# A substituição do Pandas também suporta regex
df = pd.read_csv('datasets/log.csv')
df
```

```python
# Para substituir usando regex, criamos o primeiro parâmetro para substituir o padrão regex que queremos substituir, o segundo parâmetro, o valor que queremos emitir em uma correspondência, e então passamos um terceiro parâmetro que diz apenas que regex é igual a true. 

df.replace(to_replace = '.*.html$', value = 'webpage', regex = True)
```
