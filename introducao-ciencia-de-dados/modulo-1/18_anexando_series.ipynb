# Anexando Séries

Uma última observação sobre o uso dos operadores de indexação para acessar os dados da série. O atributo .loc permite não apenas modificar os dados no local, mas também adicionar novos dados, isto é, se o valor que inserimos como índice não existir, uma nova entrada será criada, como no dicionário. 

```python
import pandas as pd 
```

```python
# Vamos a um exemplo usando uma série de algus números
s = pd.Series([1, 2, 3])
print(s)

# Vamos adicionar um novo valor
s.loc['Historia'] = 102
print(s)
```

```python
# Até agora foi mostrado apenas exemplos de uma série em que os valores do índice eram exclusivos (unique). Vamos a um exemplo em que os valores do índice não são exclusivos. 

# Vamos criar uma série de estudantes e cursos que eles fizeram
aulas_estudantes = pd.Series({'Alice': 'Física', 
                    'Jack': 'Química',
                    'Molly': 'Inglês',
                    'Sam': 'História'})

aulas_estudantes
```

```python
# Agora, vamos criar uma série só para a nova aluna Kelly, na qual lista todos os cursos que ela fez. Definiremos o índice como Kelly e os dados como os nomes dos cursos.

aulas_kelly = pd.Series(['Filosofia', 'Artes', 'Matemática'], index = ['Kelly', 'Kelly', 'Kelly'])
aulas_kelly
```

```python
# Finalmente, podemos anexar todos os dados dessa nova série à primeira usando a função .concat(). Isso criará alguma variável em todas as classes dos alunos.

todos_estudantes = pd.concat([aulas_estudantes, aulas_kelly])

# Isso cria uma série que inclui nosso pessoal original bem como todos os cursos da Kelly. 
todos_estudantes
```

Isso apenas juntou as duas séries em uma nova sem alterar as antigas

```python
# Agora, como não temos mais um index exclusivo (Kelly), podemos ver que quando consultamos a série anexada para o index Kelly não teremos mais um único valor, teremos uma nova série.

todos_estudantes.loc['Kelly']
```
